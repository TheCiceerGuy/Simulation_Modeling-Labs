```python
import random
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Simulate 10,000 dice rolls
num_rolls = 10000
results = [random.randint(1, 6) for _ in range(num_rolls)]

# Step 2: Compute frequency of each outcome
outcomes, counts = np.unique(results, return_counts=True)

# Step 3: Plot histogram
plt.bar(outcomes, counts, color='blue', alpha=0.7)
plt.xlabel("Dice Face")
plt.ylabel("Frequency")
plt.title("Dice Roll Simulation (10,000 rolls)-Shishir Ghimire")
plt.xticks(outcomes)
plt.show()

# Step 4: Print expected vs. observed probabilities
expected_prob = 1/6
observed_prob = counts / num_rolls

for i in range(6):
    print(f"Face {i+1}: Expected = {expected_prob:.4f}, Observed = {observed_prob[i]:.4f}")
```
